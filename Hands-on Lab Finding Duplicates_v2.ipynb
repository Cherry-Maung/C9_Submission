### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
df.columns
```

```python
## Write your code here
duplicate_counts = df.groupby(['MainBranch', 'Employment', 'RemoteWork']).size().reset_index(name='Count')
duplicates_only = duplicate_counts[duplicate_counts['Count']>1]
#distribution = df['Country'].value_counts().reset_index(name='Count')
duplicates_only.shape
#duplicates_only.head()
#distribution.rename(columns={'index': 'Country','Employment'}, inplace=True) # Rename column for clarity
```

```python
!pip install seaborn
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=duplicates_only, x='MainBranch', y='Count', hue='RemoteWork')
plt.title('Distribution of Duplicates by Main Branch and Remote Work')
plt.xlabel('Category')
plt.ylabel('Count of Duplicates')
plt.legend(title='RemoteWork')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='RemoteWork', order=df['RemoteWork'].value_counts().index)
plt.title('Distribution of Items Across Categories')
plt.xlabel('RemoteWork')
plt.ylabel('Count')
plt.show()
```
